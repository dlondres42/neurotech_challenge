```python
columns_1 = train.columns[train.isin(['BAIXISSIMA', 'BAIXA', 'MEDIA', 'ALTA', 'ALTISSIMA']).any()]
columns_2 = train.columns[train.isin(['MUITO PROXIMO', 'PROXIMO', 'MEDIO', 'LONGE', 'MUITO LONGE']).any()]
columns_3 = train.columns[train.isin(['E', 'D', 'C', 'B', 'A']).any()]
```

```python
encoder = OrdinalEncoder(categories=[['BAIXISSIMA', 'BAIXA', 'MEDIA', 'ALTA', 'ALTISSIMA']] * len(columns_1))
encoder2 = OrdinalEncoder(categories=[['MUITO PROXIMO', 'PROXIMO', 'MEDIO', 'LONGE', 'MUITO LONGE']] * len(columns_2))
encoder3 = OrdinalEncoder(categories=[['E', 'D', 'C', 'B', 'A']] * len(columns_3))
imputer = RandomSampleImputer()
```

```python
train.loc[:] = imputer.fit_transform(train)
```

```python
train[columns_1] = encoder.fit_transform(train[columns_1])
train[columns_2] = encoder2.fit_transform(train[columns_2])
train[columns_3] = encoder3.fit_transform(train[columns_3])
```

```python
train_numeric = train.select_dtypes(include='number')
train_numeric = train_numeric.iloc[:, :-1]
train_numeric
```

```python
special_columns = train_numeric.columns
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import classification_report, roc_auc_score
```

```python
train_numeric['TARGET'].mean()
```

```python
X = train_numeric.iloc[:, 1:]
y = train_numeric['TARGET']

split = StratifiedShuffleSplit(n_splits=1, test_size=0.02, random_state=42)

for train_index, test_index in split.split(X, y):
    X_train = X.loc[train_index]
    y_train = y.loc[train_index]
    X_test = X.loc[test_index]
    y_test = y.loc[test_index]
    rfc = RandomForestClassifier(n_estimators=200,
                              max_depth=10,
                              min_samples_split=16,
                              class_weight='balanced',
                              max_samples=0.7,
                              random_state=69)
    
    xgb = XGBClassifier(n_estimators=200,
                              max_depth=10,
                              min_child_weight=16,
                              scale_pos_weight=0.25,
                              subsample=0.7,
                              random_state=69)
    
    rfc.fit(X_train, y_train)
    xgb.fit(X_train, y_train)
    
    y_pred_rfc = rfc.predict(X_test)
    y_pred_xgb = xgb.predict(X_test)
    
    y_probs_rfc = rfc.predict_proba(X_test)[:,1]
    y_probs_xgb = xgb.predict_proba(X_test)[:,1]
    
    roc_rfc = roc_auc_score(y_test, y_probs_rfc)
    roc_xgb = roc_auc_score(y_test, y_probs_xgb)
    
    print(roc_rfc)
    print(roc_xgb)
    print(classification_report(y_test, y_pred_rfc))
    print(classification_report(y_test, y_pred_xgb))
```

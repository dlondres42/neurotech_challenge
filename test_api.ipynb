# Requisição 2

```python
train = pd.read_csv('datasets/credit_01/train.gz', compression='gzip', header=0, sep=',', engine='python')
train.head()
```

```python
oot = pd.read_csv('datasets/credit_01/oot.gz', compression='gzip', header=0, sep=',', low_memory=False)
oot.head()
```

```python
test = pd.read_csv('test.gz', compression='gzip', header=0, sep=',', low_memory=False)
test.head()
```

```python
### Checando se não tem categorias que não apareceram no treinamento
### Se tiver substituindo por NaN (da pra usar algum outro metodo melhor)
### Troca isso pelo input da requisição e deixa só um
### Tira o tqdm, deixa só o zip
### Na coluna VAR121 tem uma categoria que não tinha no treinamento, precisa corrigir isso pra poder scorar

for var, cat in tqdm(zip(model[0].transformers_[1][1].feature_names_in_, model[0].transformers_[1][1][1].categories_), total=99):
    train[var] = train[var].apply(lambda x: np.nan if x not in cat else x)
    oot[var] = oot[var].apply(lambda x: np.nan if x not in cat else x)
    
    
```

```python
ks_2samp(model.predict_proba(train)[:, 1], model.predict_proba(test)[:, 1])
```

```python
ks_2samp(model.predict_proba(oot)[:, 1], model.predict_proba(test)[:, 1])
```
